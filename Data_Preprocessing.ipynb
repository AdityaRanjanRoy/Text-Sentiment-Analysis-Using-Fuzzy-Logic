```python
from googletrans import Translator
import pandas as pd
import time
import traceback
import tweepy
```

```python
api_key="*****************************"
api_secret="*************************"
access_token="***************************"
access_secret="****************************"

auth=tweepy.OAuthHandler(api_key,api_secret)
auth.set_access_token(access_token,access_secret)
api=tweepy.API(auth,wait_on_rate_limit=True)
```

```python
tweets,news_media=[], ['cnnbrk','France24_fr']
count=1000
try:
    for news in news_media:
        print(f"Scraping:{news}")
        for tweet in api.user_timeline(id = news, count=count):
            tweets.append((news,tweet.created_at,tweet.id,tweet.text))
        time.sleep(15)
    df=pd.DataFrame(tweets,columns=['news_media','tweet_created_at','tweet_id','text'])
    print("Completed")
    
except BaseException as e:
    traceback.print_exc()
```

```python
df.head(10)
```

```python
tweets,text_query=[],'corona'
count=1000

try:
    for tweet in api.search_tweets(q=text_query,count=count,result_type='recent',include_entities=True,monitor_rate_limit=True,wait_on_rate_limit=True):
        tweet_text=""
        try:
            tr=Translator()
            if tweet.lang and tweet.lang not in ['en','und']:
                
                if tweet.text:
                    translated=tr.translate(tweet.text)
                    if translated:
                        tweet_text=translated.__dict__()["text"]
                else:
                    tweet_text=tweet.text
        except Exception as e:
            traceback.print_exc()
            pass
        
        tweets.append((tweet.created_at,tweet.id,tweet.text,tweet_text,tweet.user.screen_name,tweet.user.name,tweet.lang))
        
        df=pd.DataFrame(tweets,columns=['created_at','tweet_id','text_row','text_en','user_name','name','language'])
        time.sleep(3)
        
    print("completed")
    
except BaseException as e:
    traceback.print_exc()
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('punkt')
stop_words=stopwords.words('engLish')
```

```python
print(stopwords.words('english'))
stop_words=stopwords.words('engLish')
```

```python

def cleaning(data):
    data=str(data).lower()
    data=re.sub(r"@\s+ ",r'',data)
    
    for emoji in emojis.keys():
        data=data.replace(emoji,emojis[emoji])
    data=re.sub("\s+",' ',data)
    data=re.sub("\n",' ',data)
    letters=re.sub("[^a-zA-Z]"," ",data)
    
    return letters

def stopword(words):
    #return ",".join([word for word in str(x).split() if word not in stop_words])
    filter_words=[]
    for w in words:
        if w not in stop_words:
            filter_words.append(w)
    return filter_words

```

```python
#clean the data
af['text_en']=af['text_en'].apply(lambda x:cleaning(x))
af.head(5)
```

```python
#remove stop words
af['text_en']=af['text_en'].apply(lambda x:word_tokenize(x))
af['text_en']=af['text_en'].apply(lambda x:stopword(x))
```

```python
af['text_en']=af['text_en'].apply(lambda x: ' '.join(x))
af.head(5)
```
